## Broadcasting Operations

```python
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba["Age"].add(5)
nba["Age"] + 5

nba["Salary"].sub(5000000)
nba["Salary"] - 5000000

nba["Weight"].mul(0.453592)
nba["Weight in Kg"] = nba["Weight"] * 0.453592
```

```python
nba.head()
```

```python
nba["Salary"].div(1000000)
nba["Salary in Million"] = nba["Salary"] / 1000000
```

## The .astype() Method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None", inplace=True)
nba.head()
```

```python
# This method require a Series to not have a null values

nba.dtypes
```

```python
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.info()
```

```python
nba
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
```

```python
nba
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
# Look at the usage memory, it's been reduced
nba.info()
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")
```

```python
nba.info()
```
